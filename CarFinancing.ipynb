Reduce the dataset to core values by group by

Assumption: The Starting Balance is assumed to be the cost of the car

```python
df3 = df2.groupby(['car_type','interest_rate']).max('starting_balance')
```

```python
df3
```

```python
df4 = df3.reset_index()
```

```python
df4
```

Cleaning the dataset

```python
df4.rename(columns = {'Starting Balance':'Cost'}, inplace = True)
```

```python
df5 = df4.drop(columns=['Month', 'Repayment','Interest Paid','Principal Paid', 'New Balance'])
```
